```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import matplotlib.colors as mcolors
import matplotlib.dates as mdates
import matplotlib.dates as mdates
from matplotlib.ticker import MaxNLocator
import re
```

```python
# Load the uploaded Excel file to check its structure
file_path = '/Users/hongjiewang/Documents/Dragon Boat/DragonBoat_PKU/2025-03-17至2025-05-11_打卡记录.xlsx'
xls = pd.ExcelFile(file_path)
xls.sheet_names
```

```python


# 准备统计总分
total_scores = {}

# 遍历7天的打卡记录
for sheet in xls.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet)
    # 保留姓名和打卡类型两列
    df = df[['姓名', '打卡类型']].dropna(subset=['打卡类型'])
    
    # 提取打卡类型中的分数字段
    for _, row in df.iterrows():
        name = row['姓名']
        match = re.search(r'\+(\d+)', str(row['打卡类型']))
        score = int(match.group(1)) if match else 0
        total_scores[name] = total_scores.get(name, 0) + score

# # 转为DataFrame，方便后续处理
score_df = pd.DataFrame(list(total_scores.items()), columns=['姓名', '总分']).sort_values(by='总分', ascending=False).reset_index(drop=True)
# import ace_tools as tools; tools.display_dataframe_to_user(name="每个人7天打卡总分", dataframe=score_df)

score_df
```

```python
plt.rcParams['font.family'] = 'Heiti TC'  # 黑体，支持中文

# 图1：柱状图 - 总分排名
plt.figure(figsize=(8, 6))
# plt.bar(score_df['姓名'], score_df['总分'], color='green', edgecolor='black')
sns.barplot(x='姓名', y='总分', data=score_df, palette='spring', edgecolor='black') 
plt.xlabel('姓名', fontsize=14)
plt.ylabel('总分', fontsize=14)
plt.title(f'双周打卡总分{xls.sheet_names[0]} - {xls.sheet_names[-1]}', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.yticks(fontsize=12)
#plt.savefig(f'../data/figures/cumulative-ranking-workout-check-in-{xls.sheet_names[-1]}.png')
plt.show()

```

```python
# 用完整sheet名对应日期
# day_mapping = {
#     '2025年03月17日': '第1天',
#     '2025年03月18日': '第2天',
#     '2025年03月19日': '第3天',
#     '2025年03月20日': '第4天',
#     '2025年03月21日': '第5天',
#     '2025年03月22日': '第6天',
#     '2025年03月23日': '第7天',
#     '2025年03月24日': '第8天',
#     '2025年03月25日': '第9天',
#     '2025年03月26日': '第10天',
#     '2025年03月27日': '第11天',
#     '2025年03月28日': '第12天',
#     '2025年03月29日': '第13天',
#     '2025年03月30日': '第14天',
# }

# day_mapping = {
#     '2025年04月14日': '4月14日',
#     '2025年04月15日': '4月15日',
#     '2025年04月16日': '4月16日',
#     '2025年04月17日': '4月17日',
#     '2025年04月18日': '4月18日',
#     '2025年04月19日': '4月19日',
#     '2025年04月20日': '4月20日',
#     '2025年04月21日': '4月21日',
#     '2025年04月22日': '4月22日',
#     '2025年04月23日': '4月23日',
#     '2025年04月24日': '4月24日',
#     '2025年04月25日': '4月25日',
#     '2025年04月26日': '4月26日',
#     '2025年04月27日': '4月27日',
# }

# create an automatic day mapping
nth_day_mapping = {}
back_to_date_mapping = {}
for i, sheet in enumerate(xls.sheet_names):
    # nth_day_mapping[sheet] = f'第{i+1}天'
    nth_day_mapping[sheet] = f'{i+1}-day'
    back_to_date_mapping[f'{i+1}-day'] = f'{sheet}'
day_mapping = nth_day_mapping

# 初始化每日得分DataFrame
daily_scores = {}

for sheet in xls.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet)
    df = df[['姓名', '打卡类型']].dropna(subset=['打卡类型'])

    # 提取分数
    day_score = df.copy()
    day_score['得分'] = df['打卡类型'].str.extract(r'\+(\d+)').astype(float).fillna(0)
    daily_sum = day_score.groupby('姓名')['得分'].sum().reset_index()
    daily_sum = daily_sum.rename(columns={'得分': day_mapping[sheet]})
    daily_scores[day_mapping[sheet]] = daily_sum

# 合并所有天的分数
all_days = list(day_mapping.values())
final_df = daily_scores[all_days[0]]
for d in all_days[1:]:
    final_df = pd.merge(final_df, daily_scores[d], on='姓名', how='outer')

# 填充空值为0，并计算总分
final_df = final_df.fillna(0)
score_columns = [col for col in final_df.columns if col in day_mapping.values()]
final_df['总分'] = final_df[score_columns].sum(axis=1)
final_df = final_df.sort_values(by='总分', ascending=False).reset_index(drop=True)
final_df.index += 1  # 排名从1开始
final_df.reset_index(inplace=True)
final_df = final_df.rename(columns={'index': '排名'})


final_df
# import ace_tools as tools; tools.display_dataframe_to_user(name="打卡总榜（含第1天、第2天、第7天和总分）", dataframe=final_df[['排名', '姓名', '第1天', '第2天', '第7天', '总分']])
```

```python
# Make a cumulative score DataFrame
df_accumulative = final_df.copy()
for i in range(1, len(score_columns)):
    df_accumulative[score_columns[i]] = df_accumulative[score_columns[i-1]]+ df_accumulative[score_columns[i]]
df_accumulative 
```

```python
# plot a line for each row, with a different color for each line, for the scores_columns in df_accumulative,
# and the x-axis as the score_columns
plt.figure(figsize=(8, 6))
# Set color palette
cmap = cm.get_cmap('tab20b', len(df_accumulative))
colors = [mcolors.to_hex(cmap(i)) for i in range(len(df_accumulative))]
# Plot each line
for i, row in df_accumulative.iterrows():
    plt.plot(score_columns, row[score_columns], marker='o', color=colors[i], label=row['姓名'])
xticks = list(back_to_date_mapping.values())
# plt.xticks(rotation=90)
plt.xticks(range(len(xticks)), xticks, rotation=90)
plt.xlabel('打卡日期', fontsize=16, labelpad=10)
plt.ylabel('累计分数', fontsize=16, labelpad=10)
plt.title(f'打卡分数变化趋势 {xticks[0]} - {xticks[-1]}', fontsize=18, pad=10)
# put the legend outside the plot and make them in two columns
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=2)
# plt.savefig(f'../data/figures/daily-workout-checkin-results-by-{xticks[-1]}.png', bbox_inches='tight')
```
